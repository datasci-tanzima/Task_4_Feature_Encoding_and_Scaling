```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("/content/adult.csv")
df.head()
```

```python
df.info()
```

* Numerical: age, fnlwgt, education-num, capital-gain, capital-loss, hours-per-week
* Categorical: workclass, education, marital-status, occupation, relationship, race, sex, native-country
* Target: income

```python
le = LabelEncoder()
df['income'] = le.fit_transform(df['income'])
```

```python
df = pd.get_dummies(df, drop_first=True)
df.head()
```

```python
scaler = StandardScaler()

numerical_cols = ['age', 'fnlwgt', 'educational-num',
                  'capital-gain', 'capital-loss', 'hours-per-week']

df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

* Before scaling: numerical values had different ranges
* After scaling: features have mean ≈ 0 and standard deviation ≈ 1
Dataset is now model-ready

```python
df.to_csv("adult_income_preprocessed.csv", index=False)
```
